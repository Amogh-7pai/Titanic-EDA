# **Exploratory Data Analysis (EDA) on Titanic Dataset**

In this notebook, we will analyze the preprocessed Titanic dataset to uncover patterns, relationships, and insights that inform future modeling.

### 1. Data Loading & Initial Checks

Goal: Verify the preprocessed data is ready for EDA and understand its structure.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

# Load the preprocessed dataset (fare is log transformed)
df = pd.read_csv('data/Titanic_dataset_new.csv')  

# Check basic info
print(df.info())
print("\nMissing values:\n", df.isnull().sum())

# Display first 5 rows
df.head()
```

### 2. Statistical Summaries

Goal: Understand distributions and central tendencies of key features.

```python
# Summary statistics for numerical features
print(df.describe())

# For binary features (e.g., Survived, Sex_male)
print(df['Survived'].value_counts(normalize=True))
```

### 3. Target Variable Analysis (Survived)

Goal: Explore how survival relates to other features.

```python
# Survival rate by passenger class
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title('Survival Count by Passenger Class')
plt.show()

# Survival rate by gender
sns.barplot(x='Sex_male', y='Survived', data=df)
plt.title('Survival Rate by Gender (1=Male)')
plt.show()
```

### 4.. Feature Distributions & Relationships

Goal: Identify patterns and correlations between features.

```python
# Histograms for numerical features
df[['Age', 'Fare','SibSp','Parch']].hist(bins=20, figsize=(10, 6))
plt.suptitle('Distributions) ')

# Correlation matrix
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Feature Correlation Matrix')
plt.show()

# Family size analysis
df['FamilySize'] = df['SibSp'] + df['Parch']
sns.countplot(x='FamilySize', hue='Survived', data=df)
plt.title('Survival by Family Size')
plt.show()

# Boxplots for numeric features by survival
plt.figure(figsize=(10,6))
sns.boxplot(x='Survived', y='Age', data=df)
plt.title('Age Distribution by Survival')
plt.show()
```

### 5. Advanced Visualizations

Goal: Dive deeper into multivariate relationships.

```python
# Convert Fare back to original scale for interpretability 
df['Fare_original'] = np.exp(df['Fare'])  

# Select key features and plot
features = ['Age', 'Fare_original', 'Pclass', 'Survived']
sns.pairplot(df[features], hue='Survived', 
             plot_kws={'alpha': 0.6, 's': 30, 'edgecolor': 'k'},
             diag_kind='kde')
plt.suptitle('Pairplot of Key Features by Survival', y=1.02)
plt.tight_layout()
plt.show()
```

### 6. Insights & Conclusions

Goal: Summarize actionable findings for stakeholders.

### Key Insights:
1. **Demographics**:  
   - Survival rate: **38.4%** (higher for females and 1st-class passengers).  
2. **Feature Importance**:  
   - `Pclass` and `Sex_male` strongly predict survival.where `Sex_male` has a **-0.54** correlation with survival.  
3. **Data Quality**:  
   - No missing values; distributions are clean post-scaling.
